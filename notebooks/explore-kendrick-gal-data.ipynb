```python
import h5py

import numpy as np
import matplotlib.pyplot as plt
from pixell import enmap, enplot

from fnl_pipe.galaxy import GalCat, NullCut, LRGSouthCut, ZerrCut, AndCut

map_path = '/home/aroman/data/act_pub/act_planck_dr5.01_s08s18_AA_f090_daynight_map_srcfree.fits'
rand_cat_path = '/home/aroman/data/kendrick/02_pixellized_lss_maps/random_catalog.h5'

# with h5py.File(rand_cat_path, 'r') as file:
#     print(file.keys())

ref_map = enmap.read_map(map_path)[0]

gc = GalCat(rand_cat_path)
cut = NullCut()
gp = gc.get_subcat(cut, ref_map, vr_field='vr')
gp.import_data()

def plot_footprint():
    gal_map = gp.get_gal_map()
    fig = enplot.plot(enmap.downgrade(gal_map, 16), ticks=15)
    enplot.show(fig)

plot_footprint()
```
